```python
import pandas as pd


# Create a boolean mask to select rows where REQUESTED_DELIVERY_DATE is not zero
mask = df['REQUESTED_DELIVERY_DATE'] != 0

# Filter the DataFrame to keep only the rows where the mask is True
df = df[mask]


# Convert DELIVERY_DATE to datetime
df['DELIVERY_DATE'] = pd.to_datetime(df['DELIVERY_DATE'], format = '%Y%m%d')

potato = df['FIRST_GR_POSTING_DATE'] != 0
df = df[potato]

# Convert REQUESTED_DELIVERY_DATE to datetime
df['REQUESTED_DELIVERY_DATE'] = pd.to_datetime(df['REQUESTED_DELIVERY_DATE'], format = '%Y%m%d')

# Convert BI_LAST_UPDATED_PURCHASE_DOCUMENT_ITEM to datetime
# df['BI_LAST_UPDATED_PURCHASE_DOCUMENT_ITEM'] = pd.to_datetime(df['BI_LAST_UPDATED_PURCHASE_DOCUMENT_ITEM'], unit='s')


df.head()
```

```python
# Convert column to datetime with error handling
df['POR_DELIVERY_DATE'] = pd.to_datetime(df['POR_DELIVERY_DATE'],format = '%Y%m%d')
df['FIRST_GR_POSTING_DATE'] = pd.to_datetime(df['FIRST_GR_POSTING_DATE'],format = '%Y%m%d')
df.head()
df.info()
```

```python
df['INBOUND_DELIVERY_ITEM_ID'] = pd.to_numeric(df['INBOUND_DELIVERY_ITEM_ID'], errors='coerce').astype('Int64')
df['PLANNED_DELIVERY_DAYS'] = pd.to_numeric(df['PLANNED_DELIVERY_DAYS'], errors='coerce').astype('Int64')
df['SUB_COMMODITY_DESC'] = df['SUB_COMMODITY_DESC'].astype('category')

df.head()
```

```python

# Assuming df is your DataFrame and 'Postal Code' is the name of the column containing postal codes
# Replace 'df' and 'Postal Code' with your actual DataFrame name and column name

# Function to extract the area code from a postal code
def extract_area_code(postal_code):
    return postal_code.split('-')[0] if '-' in postal_code else postal_code

# Apply the function to the postal code column
df['AREA_CD'] = df['POSTAL_CD'].apply(extract_area_code)

# Now df will have a new column 'Area Code' containing the area codes
```

```python
df['COMPANY_CODE_ID'] = df['COMPANY_CODE_ID'].astype('category')
df['VENDOR_ID'] = df['VENDOR_ID'].astype('category')
df['AREA_CD'] = df['AREA_CD'].astype('category')
df['MATERIAL_ID'] = df['MATERIAL_ID'].astype('category')
df['PURCHASE_DOCUMENT_ID'] = df['PURCHASE_DOCUMENT_ID'].astype('category')
df['PURCHASE_DOCUMENT_ITEM_ID'] = df['PURCHASE_DOCUMENT_ITEM_ID'].astype('category')
df['PLANT_ID'] = df['PLANT_ID'].astype('category')
df['INBOUND_DELIVERY_ITEM_ID'] = df['INBOUND_DELIVERY_ITEM_ID'].astype('category')
df['MRP_TYPE_ID'] = df['MRP_TYPE_ID'].astype('category')
df.head()
```
